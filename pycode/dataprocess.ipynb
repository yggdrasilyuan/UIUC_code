```python
import numpy as np
import pandas as pd
import os

file_path="E:\\554 PROJECT2\SECOND_ETF_DATA"
files=os.listdir(file_path)
# dataname=files.split(".")[0]
# data=pd.read_csv('D:/CSV/CLOSEPRICE.csv')
# data
name=[]
for i in files:
    # print(i.split(".")[0])
    name.append(i.split(".")[0])

high=pd.read_csv("E://554 PROJECT2/SECOND_ETF_DATA/"+name[0]+".csv")
low=pd.read_csv("E://554 PROJECT2/SECOND_ETF_DATA/"+name[1]+".csv")
open=pd.read_csv("E://554 PROJECT2/SECOND_ETF_DATA/"+name[2]+".csv")
tradevolume=pd.read_csv("E://554 PROJECT2/SECOND_ETF_DATA/"+name[3]+".csv")
volume=pd.read_csv("E://554 PROJECT2/SECOND_ETF_DATA/"+name[4]+".csv")
close=pd.read_csv("E://554 PROJECT2/SECOND_ETF_DATA/"+name[5]+".csv")

len(volume.columns)
volume.columns
name
```

```python
com_name=close.columns.to_list()
for i in range(28):
    save_data=close[[com_name[2*i],com_name[2*i+1]]]
    save_data["high"]=high[[com_name[2*i+1]]]
    save_data["low"]=low[[com_name[2*i+1]]]
    save_data["open"]=open[[com_name[2*i+1]]]
    save_data["amount"]=tradevolume[[com_name[2*i+1]]]
    save_data["volume"]=volume[[com_name[2*i+1]]]
    save_data=save_data.rename(columns={com_name[2*i+1]:"close"})
    save_data=save_data.rename(columns={com_name[2*i]:"date"})
    save_data.to_csv("D://company/"+com_name[2*i]+".csv",sep=",",index=False,header=True)

# for i in range(54)
# save_data_high

# save_data=pd.concat([save_data_close,save_data_high],axis=1)
# save_data
```
