```python
import pandas as pd
import numpy as np
from scipy.optimize import root_scalar,root

def solver(time,bp_price,bond_rate):
    tmp3 = pd.DataFrame({"t": time, "cds_fee": bp_price,"bond_rate":bond_rate})
    t=tmp3["t"].to_numpy()
    phi = tmp3["cds_fee"].to_numpy()
    Btt = tmp3["bond_rate"].to_numpy()
    # phi = 0.0001 * price
    dt = 0.25
    rr=0.6

    def f1(x):
        return phi[0] - ((1 - rr) * Btt[0] * x) / (Btt[0] * (1 - (x * dt)))

    htt = [root_scalar(f1, bracket=[0, 1]).root]

    for i in range(1, len(t)):
        def f(x):
            numerator = ((1 - rr) * (sum(Btt[:i] * htt) + Btt[i] * x))
            denominator = sum([Btt[j] * (1 - sum(htt[:j+1]) * dt) for j in range(i)]) + Btt[i] * (1 - (sum(htt) * dt + x * dt))
            return phi[i] - numerator / denominator

        htt.append(root_scalar(f, bracket=[-1, 1]).root)

    return htt

time=data["duration"]
bp_price=data["bp4"]
bond_rate=data["Bond_rate"]
bp_price1=data["bp5"]
# hr_f=solver(time,bp_price,bond_rate)
hr_f5=solver(time,bp_price1,bond_rate)
# hr_f=np.array(hr_f)
hr_f5=np.array(hr_f5)
# type(hr_f)
```
